```python
# Primeiro passo: Importar as duas bibliotecas necessárias para gerar os gráficos desejados
import numpy as np
import matplotlib.pyplot as plt

# Segundo passo: Definir as variáveis arbitrárias: h e raio de base
print('Olá, espero que esteja bem. Para gerar os gráficos referentes ao Came, irei precisar da sua ajuda para dizer as variáveis do seu problema, ok?')
h = float(input("Digite o valor de h: ")) # Pergunta ao usuário qual h deve ser utilizado. Exemplo: 1
raio = float(input("Digite o valor do raio de base: ")) # Pergunta ao usuário qual raio deve ser utilizado. Exemplo: 2

# Terceiro passo: Gerar os valores de lambda
lambdaa = np.linspace(0, 2*np.pi, 1000)

# Quarto passo: Iniciar os arrays para r e deslocamento
r = np.zeros_like(lambdaa)
deslocamento = np.zeros_like(lambdaa)

# Quinto passo: Definir r e deslocamento conforme o que foi solicitado
for i, angulo in enumerate(lambdaa):
    if angulo <= np.radians(90):  # De 0 a 90 graus
        r[i] = raio + h
        deslocamento[i] = h
    elif angulo > np.radians(90) and angulo <= np.radians(180): # De 91 a 180 graus
        r[i] =  raio + 2*h
        deslocamento[i] = 2*h
    else:
        r[i] = raio
        deslocamento[i] = 0

# Quinto passo: Criar o gráfico de linha
plt.plot(np.degrees(lambdaa), deslocamento)
plt.xlabel('Ângulo do cames (λ)')
plt.ylabel('Deslocamento (s)')
plt.title('Deslocamento (s) em função de Ângulo (λ)')
plt.show()

# Sexto passo: Criar o gráfico polar
plt.figure(figsize=(6, 6))
plt.subplot(111, polar=True)
plt.plot(lambdaa, r)
plt.title('Gráfico Polar da Geometria do Came', va='bottom')
plt.grid(True)
plt.show()

```
